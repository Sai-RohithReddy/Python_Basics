```python
from matplotlib import pyplot as plt

x=[10,20,30,40,50]
y=[100,200,300,400,500]

plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.xticks(x,['a','b','c','d','e'])
plt.yticks(y,['0-100','100-200','200-300','300-400','400-500'])

plt.show()
```

```python
from matplotlib import pyplot as plt

x=[10,20,30,40,50]
y=[100,200,300,400,500]

plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.xlim(0,100)
plt.ylim(0,1000)

plt.show()
```
