# BAR PLOT
it is used to represent the category of data with rectangular bars with length and height which is proportional to the curresponding values

```python
import matplotlib.pyplot as plt
```

```python
x=['A','B','C','D']
y=[10,30,40,20]

plt.bar(x,y)
plt.show()
```

```python
x=['A','B','C','D']
y=[10,30,40,20]

plt.bar(x,y,color='red',width=0.2)#default width=0.8
plt.show()
```

```python
x=['A','B','C','D']
y=[10,30,40,20]

plt.barh(x,y,color='b',height=0.2)#default width=0.8
plt.show()
```

```python
x=['A','B','C','D']
y=[10,30,40,20]
c=['red','green','blue','black']

plt.barh(x,y,color=c,height=0.5)#default width=0.8
plt.show()
```

```python
x=['2017','2018','2019','2020']
y=[236,530,255,489]
plt.bar(x,y)
plt.xlabel('YEARS')
plt.ylabel('PLACED')
plt.title('CANDIDATES PLACEMENT DETAILS')
plt.show()
```
