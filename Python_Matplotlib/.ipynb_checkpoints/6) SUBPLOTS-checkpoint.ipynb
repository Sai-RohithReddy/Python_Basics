# SUBPLOT
is used to draw multiple plots in a single figure

```python
import matplotlib.pyplot as plt
```

```python
x=[0,1,2,3,4]
y1=[50,20,30,40,10]
y2=[40,20,50,20,10]

plt.subplot(1,2,1) #(r,c,i)
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.plot(x,y2)

plt.show()
```

```python
x=[0,1,2,3,4]
y1=[50,20,30,40,10]
y2=[40,20,50,20,10]
y3=[5,2,3,4,1]
y4=[50,20,10,20,40]

plt.suptitle('MULTIPLE PLOTS')

plt.subplot(2,2,1) #(r,c,i)
plt.title('plot-1')
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.title('plot-2')
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.title('plot-3')
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.title('plot-4')
plt.plot(x,y4)

plt.show()
```

```python
x=[0,1,2,3,4]
y1=[50,20,30,40,10]
y2=[40,20,50,20,10]
y3=[5,2,3,4,1]

plt.subplot(3,1,1) #(r,c,i)
plt.plot(x,y1)

plt.subplot(3,1,2)
plt.plot(x,y2)

plt.subplot(3,1,3)
plt.plot(x,y3)

plt.show()
```
