# STACKED
it means placing one above the other

```python
from matplotlib import pyplot as plt
```

```python
years=['2017','2018','2019','2020']
placements=[150,275,324,295]
cse=[75,80,47,82]
it=[46,58,35,52]
ece=[67,59,72,62]
plt.title('PLACE MENT COMPARISION')
plt.bar(years,placements)
plt.bar(years,cse)
plt.bar(years,ece)
plt.bar(years,it)

plt.ylim(0,400)
plt.show()
```

```python
years=['2017','2018','2019','2020']
cse=[75,80,47,82]
it=[46,58,35,52]
ece=[67,59,72,62]

w=0.4
ece_start=[cse[i]+it[i] for i in range(len(cse))]
plt.title('PLACE MENT COMPARISION')
plt.bar(years,cse,color='black',width=w)
plt.bar(years,it,bottom=cse,color='green',width=w)
plt.bar(years,ece,bottom=ece_start,color='red',width=w)

plt.ylim(0,250)
plt.show()
```

```python
years=['2017','2018','2019','2020']
cse=[75,80,47,82]
it=[46,58,35,52]
ece=[67,59,72,62]
eee=[53,47,36,52]

ece_start=[cse[i]+it[i] for i in range(len(cse))]
eee_start=[ece[i]+ece_start[i] for i in range(len(ece))]

h=0.4
plt.title('PLACE MENT COMPARISION')
plt.barh(years,cse,color='black',height=h,label='CSE')
plt.barh(years,it,left=cse,color='green',height=h,label='IT')
plt.barh(years,ece,left=ece_start,color='red',height=h,label='ECE')
plt.barh(years,ece,left=eee_start,color='b',height=h,label='EEE')
plt.legend()

plt.xlim(0,350)
plt.show()
```
