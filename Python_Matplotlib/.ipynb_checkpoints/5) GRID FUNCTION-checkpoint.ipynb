```python
from matplotlib import pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)
plt.grid()#axis='both' -by default
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)
plt.grid(axis='y')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)
plt.grid(axis='both', ls='dotted', lw=1, c='r')
plt.show()
```
