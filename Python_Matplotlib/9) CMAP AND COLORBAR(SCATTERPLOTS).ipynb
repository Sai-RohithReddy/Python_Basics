```python
from matplotlib import pyplot as plt
```

```python
x = [10,20,30,40,50]
y = [200,300,100,500,400]
colors=['red','green','blue','black','yellow']
size=[100,30,200,250,150]

plt.scatter(x,y,c=colors,s=size)
plt.show()
```

```python
x = [10,20,30,40,50]
y = [200,300,100,500,400]
colors=[70,20,80,100,50]
size=[100,30,200,250,150]

plt.scatter(x,y,c=colors,s=size,cmap='viridis')
# cmap=Accent, BuGn, Dark2, Greys, Oranges, Purples, Reds, Spectral, binary, copper, magma, pink, rainbow, viridis
plt.colorbar()
plt.show()
```
