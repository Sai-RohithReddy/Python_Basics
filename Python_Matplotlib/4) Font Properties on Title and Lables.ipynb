```python
from matplotlib import pyplot as plt

x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)

f1 = {'family':'cambria','color':'r', 'size':20}
plt.title('Font Properties',fontdict=f1)

plt.xlabel('X-axis',fontdict={'family':'arial','color':'b', 'size':15})
plt.ylabel('Y-axis',fontdict={'family':'arial','color':'b', 'size':15})

plt.show()
```
