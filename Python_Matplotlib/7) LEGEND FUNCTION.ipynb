# Legend
are used to describe the functionality of each and every line
* legend() -> without parameters
* legend(parameters) -> with parameters


```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(0,5,1)
y1=x**2
y2=x**3

# plt.plot(x,y1,x,y2)
'''with parameter'''
# plt.legend(['Squares','Cubes'])

plt.plot(x,y1,label='Squares')
plt.plot(x,y2,label='Cubes')
# legend() calling is mandatory
plt.legend(
    loc='upper center',
    framealpha=0.5,
    facecolor='yellow',
    edgecolor='black',
    fancybox=True,
    shadow=True,
    fontsize=15
)

plt.show()
```

```python
x = np.arange(0,5,1)
y1=x**2
y2=x**3

plt.plot(x,y1,x,y2)
'''with parameter'''
plt.legend(['Squares','Cubes'])

plt.show()
```
