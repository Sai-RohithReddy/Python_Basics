```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
years=['2017','2018','2019','2020']
cse=[75,80,47,82]
it=[46,58,35,52]
ece=[67,59,72,62]
eee=[53,47,36,52]

plt.figure(figsize=(10,7))# it has to be at top

plt.title("PLACEMENT COMPARISION")

w=0.2

cse_bar=np.arange(len(years))
print(cse_bar)
plt.bar(cse_bar,cse,width=w,label='CSE')

it_bar=[i+w for i in cse_bar]
print(it_bar)
plt.bar(it_bar,it,width=w,label='IT')

ece_bar=[i+w for i in it_bar]
print(ece_bar)
plt.bar(ece_bar,ece,width=w,label='ECE')

eee_bar=[i+w for i in ece_bar]
print(eee_bar)
plt.bar(eee_bar,eee,width=w,label='EEE')

plt.xticks(cse_bar+w+w/2,years)#used to bring the years to middle

plt.xlabel('YEARS')
plt.ylabel('Placement Count')
plt.ylim(0,100)
plt.legend()
plt.show()
```

```python
years=['2017','2018','2019','2020']
cse=[75,80,47,82]
it=[46,58,35,52]
ece=[67,59,72,62]
eee=[53,47,36,52]

plt.figure(figsize=(10,7))# it has to be at top

plt.title("PLACEMENT COMPARISION")

l=0.2

cse_bar=np.arange(len(years))
print(cse_bar)
plt.barh(cse_bar,cse,height=l,label='CSE')

it_bar=[i+l for i in cse_bar]
print(it_bar)
plt.barh(it_bar,it,height=l,label='IT')

ece_bar=[i+l for i in it_bar]
print(ece_bar)
plt.barh(ece_bar,ece,height=l,label='ECE')

eee_bar=[i+l for i in ece_bar]
print(eee_bar)
plt.barh(eee_bar,eee,height=l,label='EEE')

plt.yticks(cse_bar+w+w/2,years)#used to bring the years to middle

plt.xlabel('Placement Count')
plt.ylabel('YEARS')
plt.xlim(0,100)
plt.legend()
plt.show()
```
