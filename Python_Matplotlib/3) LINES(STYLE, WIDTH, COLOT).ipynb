# line style
* ls = 'solid' / 'dotted' / 'dashed' / 'dashdot' / 'none'

# line width
* lw = "15"

# line color
* c = 'r'

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x, y, ls = 'dashed', c = 'r', lw = '5')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x, y, ls='None', marker='o')
plt.show()
```
