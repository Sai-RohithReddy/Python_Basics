```python
from matplotlib import pyplot as plt
```

```python
student_performance=['Excellent','Good','Average','Poor']
student_values=[15,25,12,8]

plt.pie(student_values)
plt.show()
```

```python
student_performance=['Excellent','Good','Average','Poor']
student_values=[15,25,12,8]

plt.pie(student_values,labels=student_performance,startangle=90,autopct='%2.1f%%')
plt.show()
```

```python
student_performance=['Excellent','Good','Average','Poor']
student_values=[15,25,12,8]

plt.figure(figsize=(8,10))

plt.pie(student_values,labels=student_performance,explode=[0.2,0,0.3,0],shadow=True,colors=['black','blue','yellow','green'])

plt.legend(title='Performances')
plt.show()
```
