```python
import pandas as pd
```

```python
d = pd.read_excel('GRADES.xlsx')
df = pd.DataFrame(d)
print(df)
```

# LOC is used to retrive the data in rows and columns, and it is completely labled based data selectin method

```python
# LOC[ROW,COLUMN]

df.loc[3] #it will print 3rd row
```

```python
df.loc[3,'Name_Of_Student'] #it will print 3rd rows name
```

```python
df.loc[0:4]
```

```python
df.loc[0:4,'Name_Of_Student']
```

```python
df.loc[0:4,['Name_Of_Student','Roll_NO']]
```

```python
df.loc[0:4,'Name_Of_Student':'Science']
```

# iLOC is also used to retrive the data in rows and columns, but here it is in index based data selection method

```python
df.iloc[0:5,1:6]
```

```python
df.iloc[:,1:6]
```

```python
df.iloc[[0,3,7]]
```

```python
df.iloc[:,[0,3]]
```
