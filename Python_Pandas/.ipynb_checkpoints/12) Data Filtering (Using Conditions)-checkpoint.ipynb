```python
import pandas as pd
p = pd.read_excel("GRADES.xlsx")
df = pd.DataFrame(p)
df
```

```python
# filtering using LOC()
df.loc[df['English']<60]
```

```python
df.loc[(df['English']<60) & (df['Science']<60)] # and statement will not work here, we need to use bit and, or(&,|) here
```

```python
df.loc[(df['English']<60) | (df['Science']<60)]
```

```python
# filtering using string
df.loc[df['Name_Of_Student'] == 'shyam']
```

```python
# filtering using sub-string
df.loc[df['Name_Of_Student'].str.contains('hi')]
```

```python
# filtering names which dosen't contain sub-string
df.loc[~df['Name_Of_Student'].str.contains('s')]
```

```python
df.loc[df['Name_Of_Student'].str.startswith('s')]
```

```python
df.loc[~df['Name_Of_Student'].str.startswith('s')]
```

```python
df.loc[df['Name_Of_Student'].str.endswith('s')]
```

```python
df.loc[~df['Name_Of_Student'].str.endswith('s')]
```
