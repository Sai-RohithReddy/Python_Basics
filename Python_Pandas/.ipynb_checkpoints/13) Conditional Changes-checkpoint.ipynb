```python
import pandas as pd
d = pd.read_excel('GRADES.xlsx')
df = pd.DataFrame(d)
df
```

```python
# adding total column in a Data Frame

df['Total'] = df['Teluge'] + df['English'] + df['Maths'] + df['Science'] + df['Social']
df
```

```python
# adding Percentage column in a Data Frame

df['Percentage'] = (df['Total'] / 500) * 100
df
```

```python
# Adding graged column based on Percentage

# Before doing in Add an colume with None values
df['Grade'] = 'None'
df
```

```python
"""
< 40 - Fail
>40 & <60 - Pass
>60 & <70 - First Class
>70 - Distinction
"""

df.loc[df['Percentage']<40,['Grade']] = 'FAIL'
df.loc[(df['Percentage']>40) & (df['Percentage']<60),['Grade']] = 'SECOND CLASS'
df.loc[(df['Percentage']>60) & (df['Percentage']<70),['Grade']] = 'FIRST CLASS'
df.loc[df['Percentage']>70,['Grade']] = 'DISTINCTION'
df
```
